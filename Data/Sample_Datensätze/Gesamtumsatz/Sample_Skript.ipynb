```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import secrets

df = pd.read_csv('../Sitzungsverfolgung/Sitzungsverfolgung_Sample.csv')
df['date'] = pd.to_datetime(df['date'],format='%Y-%m-%d')
df.info()
```

```python

#df_agg = df.groupby(by=[df.index.year, df.index.month])['sessions'].sum() #summieren
#df_agg 
```

```python
df.index =  pd.to_datetime(df['date'],format='%Y-%m-%d') # jahr von monat trennen
df_agg = df[['sessions', 'sess_add_cart', 'sess_co','sess_co_done']].groupby(by=[df.index.year, df.index.month]).sum()
df_agg 
```

```python
#conversionrate berechnen
#spalte anhÃ¤ngen namens cr -> beinhaltet dann eine berechnung - sess_co_done / sitzungen gerundet -> prozent

df_agg['cr %']= round((df_agg['sess_co_done'] / df_agg['sessions']*100),2)
```
