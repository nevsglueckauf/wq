# Merging mit Geodaten

```python
import pandas as pd

sess = pd.read_csv("sessions.csv")

loc = pd.read_csv("world_cities.csv")
loc = loc.fillna('')
loc.head()
```

```python
sess
```

```python
def get_geo(name:str) -> dict:
    tmp = loc[loc['name'] == name][['lat', 'lng']]
    return {'lat': tmp['lat'], 'lng': tmp['lng'] }
    
```

```python
print(get_geo("Leipzig"))
```

```python
sess['land_pg'] = sess['landing_uri'].str[0:65]
sess.columns
```

```python
sess_clean = sess[['sess_ctry', 'sess_loc', 'datum', 'sug_platform',
       'session_duration', 'pages_per_sess', 'jump_offs', 'sess_cart_add',
       'sess_co', 'sess_co_done', 'sessions', 'land_pg']]
```
